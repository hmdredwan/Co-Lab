## Assignment 1: Basic If-Else Statement

**Task:** Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.

## Assignment 2: Multiple Conditions

**Task:** Write a Python program that asks for the user's age and prints the appropriate message:
- If the user is 18 or older, print 'Adult'
- If the user is 13 or older but under 18, print 'Teenager'
- If the user is younger than 13, print 'Child'

```python
print("Enter your age : ")
age= int(input())
if age >= 18:
  print("Adult")
elif age >=13 and age <18:
  print ("Teenager")
elif age <13:
  print ("Child")
else:
  print ("Invalid Input")
```

## Assignment 1: Sum of Natural Numbers

**Task:** Write a Python function `sum_of_natural_numbers(n)` that returns the sum of all natural numbers up to `n`. Use a loop inside the function.

```python
def sum_natural_numbers(n):
  for i in range(n+1):
    sum= (n*(n+1))/2
    return sum

n=int(input("Enter a number : "))
result= sum_natural_numbers(n)
print(result)
```

## Assignment 2: Factorial of a Number

**Task:** Write a Python function `factorial(n)` that calculates and returns the factorial of `n`. Use a loop to calculate the factorial.

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```
