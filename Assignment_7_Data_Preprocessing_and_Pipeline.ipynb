```python
from sklearn.model_selection import train_test_split
# Split into Train and Test
# TODO: Split - 80% train, 20% test

X= df.drop('price',axis=1)
y= df['price']
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=20)
print(f"\nWe are training with: {len(X_train)} houses")
print(f"We are tresting with: {len(X_test)} houses")
```

```python
#Convert Yes/No to 1/0
# Check what's in our data
from sklearn.preprocessing import LabelEncoder

label_encoder= LabelEncoder()
X_train[  'mainroad']=label_encoder.fit_transform(X_train['mainroad'])
X_train[  'guestroom']=label_encoder.fit_transform(X_train['guestroom'])
X_train[  'basement']=label_encoder.fit_transform(X_train['basement'])
X_train[  'hotwaterheating']=label_encoder.fit_transform(X_train['hotwaterheating'])
X_train[ 'airconditioning']=label_encoder.fit_transform(X_train['airconditioning'])
X_train[  'prefarea']=label_encoder.fit_transform(X_train['prefarea'])
X_train[  'furnishingstatus']=label_encoder.fit_transform(X_train['furnishingstatus'])

X_test[  'mainroad']=label_encoder.fit_transform(X_test['mainroad'])
X_test[  'guestroom']=label_encoder.fit_transform(X_test['guestroom'])
X_test[  'basement']=label_encoder.fit_transform(X_test['basement'])
X_test[  'hotwaterheating']=label_encoder.fit_transform(X_test['hotwaterheating'])
X_test[ 'airconditioning']=label_encoder.fit_transform(X_test['airconditioning'])
X_test[  'prefarea']=label_encoder.fit_transform(X_test['prefarea'])
X_test[  'furnishingstatus']=label_encoder.fit_transform(X_test['furnishingstatus'])
# print("Current  values:", X_train.head())
print(X_train.columns)
# Shows: yes, no, yes, yes, no

# Models need numbers, not text!
```

```python
#Apply StandardScaler
# Area is huge (1650-13300), Bedrooms is small (1-6)!
scaler = StandardScaler()
print("Unscaled Data")
print(X_train.head())
# X_train= scaler.fit_transform(X_train['area'])
X_train_scaled = scaler.fit_transform(X_train)
print("Scaled Data")
print(X_train.head())

X_test_scaled = scaler.transform(X_test)
# X_test_scaled = scaler.transform(X_test)
```

```python
#train and test model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
print("✅ Model trained!")
train_score = model.score(X_train, y_train)
print(f"Training accuracy: {train_score:.1%}")

# Check accuracy on test data
test_score = model.score(X_test, y_test)
print(f"Test accuracy: {test_score:.1%}")
```

```python
# Predict a House Price
# New house details:
# area=5000, bedrooms=3, bathrooms=2, stories=2,
# mainroad=yes, guestroom=no, basement=yes,
# hotwaterheating=no, airconditioning=yes,
# parking=2, prefarea=yes, furnishingstatus=furnished
#your code here
import joblib
joblib.dump(model, 'my_model.pkl')
print("✅ Model saved as 'my_model.pkl'")
joblib.dump(scaler, 'my_scaler.pkl')
print("✅ Scaler saved as 'my_scaler.pkl'")

pd.DataFrame(X_train).to_csv('train_data_scaled.csv', index=False)
print("✅ Scaled data saved as 'train_data_scaled.csv'")

model = joblib.load('my_model.pkl')
scaler = joblib.load('my_scaler.pkl')

print("\nNew House information:")
new_house = [[5000,3, 2, 2,1, 0, 1, 0,1,2,1,0,]]
prediction = model.predict(new_house)
print(f"House Price is :{prediction.round(2)}")
```
